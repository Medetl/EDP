```python
from pylab import *
x = linspace(0,1.0,1000)
y = linspace(0,1.0,1000)   

def u_x_y(x,y):
    N = 8
    g = zeros ((len(x),len(y)))
    for i in range(len(y)):
        for n in range(1,N):
          a = (400*(1-(-1)^(n)))/(n*pi)
          b = (400*(1-(-1)^(n))*(2-cosh(n*pi)))/((n*pi)*sinh(n*pi))
          g[:,i] += sin(n*pi*x)*(a*cosh(n*pi*y[i])+b*sinh(n*pi*y[i]))
    return g

from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(10,10))
ax = fig.add_subplot(111, projection="3d")

X,Y = np.meshgrid(x,y)

U = u_x_y(x,y)

plt.xlabel(''); plt.ylabel(''); 
plt.title("u(x,y)")
ax.plot_surface(X, Y, U, cmap="cool"); 
```
